#Projection for 2024

````
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX # Import the SARIMAX class
import plotly.express as px

# 1. Extract relevant columns for forecasting
forecast_data = df[['Date', 'Average Fare', 'NDC Indicator', 'Origin Region']]

# 2. Create a date range for 2024
start_date_2024 = pd.to_datetime('2024-01-01')
end_date_2024 = pd.to_datetime('2024-12-01')
date_range_2024 = pd.date_range(start_date_2024, end_date_2024, freq='MS')

# 3. Prepare DataFrame for 2024 projections
data_2024 = pd.DataFrame()

# 4. Define a function for modeling and projecting (using SARIMA as an example)
def project_2024(df):
    projected_df = pd.DataFrame()

    # Iterate through NDC Indicator and Origin Region combinations
    for ndc in df['NDC Indicator'].unique():
        for region in df['Origin Region'].unique():
            # Filter data for the specific combination
            subset = df[(df['NDC Indicator'] == ndc) & (df['Origin Region'] == region)]

            # Ensure the subset has enough data points for SARIMA
            if len(subset) < 2:
                continue
# Fit SARIMA model for the specific combination
            sarima_model = SARIMAX(subset['Average Fare'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
            sarima_results = sarima_model.fit()

            # Forecast Average Fare for 2024
            sarima_forecast = sarima_results.get_forecast(steps=12)
            sarima_forecast_df = sarima_forecast.conf_int()
            sarima_forecast_df['forecast'] = sarima_forecast.predicted_mean

            # Create a temporary DataFrame for the combination
            temp_df = pd.DataFrame({
                'Date': date_range_2024,
                'Average Fare': sarima_forecast_df['forecast'].values,
                'NDC Indicator': ndc,
                'Origin Region': region
            })

            # Append to the main 2024 DataFrame
            projected_df = pd.concat([projected_df, temp_df], ignore_index=True)

    return projected_df

# 5. Perform the projection
data_2024 = project_2024(forecast_data)

# 6. Plot the projections for 2024
fig_2024 = px.line(data_2024,
                     x='Date',
                     y='Average Fare',
                     color='NDC Indicator',
                     facet_col='Origin Region',
                     title='Projected Average Fare Trend (2024)',
                     labels={'Date': 'Date', 'Average Fare': 'Average Fare'}, color_discrete_map={'GDS': 'red', 'NDC': 'blue'})


fig_2024.update_layout(
    height=800,
    width=2000,
    showlegend=True
)

fig_2024.show()
````
